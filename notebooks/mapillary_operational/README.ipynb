## First, set the configuration.yml
The 2 most important things to set are the:
1) sys_path_append 
2) base_dir

See the yaml file for comments. The general folder structure is:

```
|__base_dir/
|   |__images/ (where downloaded images go)
|   |__images_metadata/ (where metadata about the images, like lon/lat, and detections go.
|   |__output/ (where other miscellaneous output files should go)
```



## get_mapillary_metadata.ipynb

This script assumes you already have one of more image_ids from prior mapillary searches. It's set up for you to plug the ids into a list at the top of the file, but it's easy enough to modify to import from some other source.

df_metadata: one row per image_id, containing key metadata fields like lon/lat, file name, image id, etc.



df_detections: multiple rows per image (and potentially multiple rows per detection.) 




This script will save both datafraes to the images_metadata folder
